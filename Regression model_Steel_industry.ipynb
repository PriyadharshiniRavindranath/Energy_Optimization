```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Model(nn.Module):
    def __init__(self,in_features=13,h1=8,h2=9):
        super(Model,self).__init__()
        self.fc1=nn.Linear(in_features,h1)
        self.fc2=nn.Linear(h1,h2)
        self.fc3=nn.Linear(h2,1)
    def forward(self,x):
        x=F.relu(self.fc1(x))
        x=F.relu(self.fc2(x))
        x=self.fc3(x)
        
        return x
    
    
```

```python
torch.manual_seed(43)
model=Model()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r'D:\priya\Steel_industry_data.csv') 
data
```

```python
data['date'] = pd.to_datetime(data['date'])
data['hour'] = data['date'].dt.hour
data['day_of_month'] = data['date'].dt.day
data['month'] = data['date'].dt.month
data['year'] = data['date'].dt.year
data = data.drop(['date'], axis=1)
data
```

```python
X = data.drop('Usage_kWh', axis=1) 
y = data['Usage_kWh'] 
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
numeric_columns = X.select_dtypes(include=['number']).columns
X[numeric_columns] = X[numeric_columns].fillna(X[numeric_columns].mean())
```

```python
categorical_columns = X.select_dtypes(include=['object']).columns
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    label_encoders[col] = le
```

```python
X=X.values
y=y.values
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=20,random_state=42)
```

```python

scaler_x = StandardScaler()
scaler_y = StandardScaler()
X_train = scaler_x.fit_transform(X_train)
y_train = scaler_y.fit_transform(y_train.reshape(-1, 1))
X_test = scaler_x.transform(X_test)
y_test = scaler_y.transform(y_test.reshape(-1, 1))
```

```python
model.parameters
```

```python
X_train = torch.tensor(X_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
```

```python
y_train = torch.tensor(y_train, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32)
```

```python
from torch.utils.data import DataLoader,TensorDataset
dataset = TensorDataset(X_train, y_train)
dataloader = DataLoader(dataset, batch_size=64, shuffle=True)
```

```python

input_dim = X_train.shape[1]
model = Model(in_features=input_dim)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 1000
for epoch in range(epochs):
    model.train()
    for batch_X, batch_y in dataloader:
    
        outputs = model(batch_X)
        loss = criterion(outputs.squeeze(), batch_y.squeeze())

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch + 1}/{epochs}], Loss: {loss.item():.4f}')
```

```python
model.eval()
with torch.no_grad():
    predictions = model(X_test)
    mse = criterion(predictions.squeeze(), y_test.squeeze())
    print(f'Mean Squared Error on Test Set: {mse.item():.4f}')
```
